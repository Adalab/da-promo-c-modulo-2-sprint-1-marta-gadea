### Ejercicio 10
Utilizando la columna de year cread una nueva columna llamada siglo que tiene siglo-XX o siglo-XXI. Donde el año sea menor que 2000 pondremos siglo-XX y el resto de los años siglo-XXI.


```python
db_sharks3 = db_sharks2.copy() #como se trata de un cambio importante en el csv, vamos a trabajar sobre una copia
db_sharks3.head(2)
```

```python
#creamos la columna con los filtros:
siglos = [(db_sharks3['year'] < 1900), (db_sharks3['year'] < 2000), (db_sharks3['year'] >= 2000)]
valores = ['siglo-XIX', 'siglo-XX', 'siglo-XXI']

db_sharks3['siglo'] = np.select(siglos, valores)
```

```python
#comprobamos que los valores se han introducido correctamente:
db_sharks3['siglo'].unique()
```

        Vamos a comprobar que nuestra columna siglo ha filtrado correctamente los valores de la columna year. Para ello buscamos el valor mínimo de year y su posición:

```python
db_sharks3.loc[db_sharks3['year'] == db_sharks3['year'].min()]
```

        Vemos que en la columna 'siglo' en este caso aparece 'siglo-XIX', que es el valor que queríamos que apareciera. Ha funcionado :-)
