## Ejercicio 2
Para poner en práctica el read_clipboard, elegid alguna tabla que os podaís encontrar en google y usad dicho método para traer los datos al jupyter notebook. Una vez que los hayáis cargado, guardad esos datos en los siguientes formatos:  
        - .csv  
        - .excel  
        - .json  
        - .pickle  

```python
#la tabla es sobre items clave del videojuego Persona 5 Royal.
#https://megamitensei.fandom.com/wiki/List_of_Persona_5_Royal_Items
df_clipboard = pd.read_clipboard(sep = '\t')
df_clipboard.head(10)
```

Una vez cargada la tabla, guardamos los datos:

- En CSV:

```python
df_clipboard.to_csv('../pandas/tabla_persona.csv')
```

- En Excel:

```python
df_clipboard.to_excel('../pandas/tabla_persona.xlsx')
```

- En Pickle:

```python
df_clipboard.to_pickle('../pandas/tabla_persona.pkl')
```

- En JSON:

```python
df_clipboard.to_json('../pandas/tabla_persona.json')
```

- En Parquet:

```python
df_clipboard.to_parquet('../pandas/tabla_persona.parquet')
```
