### Ejercicio 1.

```python
attacks = pd.read_csv('../Pandas/attacks_1_4.csv')
attacks.head(3)
```

```python
attacks.columns
```

```python
paises = attacks['country'].value_counts()
paises = paises.reset_index()
paises.sort_values(by='country', ascending=False)
```

### Ejercicio 2.

```python
#a√±adimos el head() al final ya que de lo contrario nos ense√±a todas las filas independientemente de lo que le pidamos X desde el m√©todo .head()
#https://stackoverflow.com/a/36074520
attack_per_year = attacks.sort_values(['year'], ascending=False).groupby('year').head() 
attack_per_year.head(3)
```

### Ejercicio 3.
¬øCu√°ntos ataques hubo por a√±o y sexo? Devolved un DataFrame solo del a√±o 2000 en adelante.

```python
year2000 = attacks[attacks['year'] >= 2000]
ejercicio3 = year2000.groupby('year')['sex']

df_ejercicio3 = ejercicio3.count().reset_index()
df_ejercicio3
```

### Ejercicio 4.
Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. ¬øCu√°ntos de los ataques fueron letales por sexo? Devuelve un DataFrame.

```python
lista_paises = ['USA', 'AUSTRALIA', 'SOUTH AFRICA']
attacks_por_pais = attacks[attacks['country'].isin(lista_paises)]
```

```python
letales = np.where(attacks_por_pais['fatal_(y/n)'] == 'Y')
sexo = np.where(attacks_por_pais['sex'])
```

```python
attacks_por_pais.groupby(letales, sexo)
```

```python
df_letal = attacks_por_pais[(attacks_por_pais['fatal_(y/n)'] == 'Y')]
df_letal.head(3)
```

```python
#ataques por sexo:
df_letal['sex'].value_counts()
```
