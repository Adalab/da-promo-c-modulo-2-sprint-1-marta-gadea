```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Ejercicio 1.
Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos.

```python
df_attacks = pd.read_csv('../Visualizacion/attacks_1_4.csv')
df_attacks.head(3)
```

```python
df_attacks_2015 = df_attacks[df_attacks['year'] >= 2015]
```

```python
plt.figure(figsize=(10, 20))
sns.countplot(y=df_attacks_2015['activity'], hue=df_attacks_2015['year'])
plt.show();
```

La gráfica relaciona la actividad que se estaba realizando aal ser atacado con el año en que ocurrió. La actividad durante la cual se han dado más ataques independientemente del año ha sido el surfing, aunque fue en 2015 cuando se registraron un número muy elevado de casos con respecto a otros años.
La siguiente actividad con más incidencia de casos fue la natación, aunque en este caso el mayor número de casos se dio en 2016.
Fueron precisamente estos años (2015 y 2016) aquellos en los que más casos se han registrado de ataques de tiburón.

### Ejercicio 2.
Usando la librería seaborn, crea una gráfica de barras que muestre el número de ataques que han ocurrido cada año en función del sexo. Hacedlo solo desde el 2015.

```python
df_attacks_sexo = (df_attacks_2015.groupby(['sex', 'year'])['case_number'].count()).reset_index()
df_attacks_sexo
```

```python
plt.figure(figsize=(5, 5))
sns.barplot(data = df_attacks_sexo, x='year', y='case_number', hue='sex')
plt.show();
```

En esta gráfica de barras queda reflejado que los ataques entre 2015 y 2018 se dieron de manera mucho más evidente en hombres que en mujeres. Antes de concluir que pudiera haber una relación directa entre la posibilidad de ser atacado y el sexo, habría que determinar primero cuál fue en esos años el porcentaje de hombres y de mujeres que realizaron actividades acuáticas y subacuáticas.

### Ejercicio 3.
¿Cuántos ataques hubo por año? Seleccionad solo desde el año 1500.

```python
df_1500 = df_attacks[df_attacks['year'] >= 1500]
df_1500.head(3)
```

```python
plt.figure(figsize=(10, 30))
sns.countplot(y=df_1500['year'], data=df_1500)
plt.show();
```

Vemos como desde el 1500 hasta 1919 los ataques de tiburón eran mínimos, y que fue hacia 1945 cuando empezó a apreciarse una ligera subida que ya en 1960 se hizo evidente. Es razonable pensar que en los primeros años reflejados por el gráfico las actividades acuáticas eran mínimas y la relación con el mar era sobretodo con fines de pesca y transporte.
A partir de 2002 y hasta alcanzar un máximo en 2015 los registros de ataques fueron muy elevados.

### Ejercicio 4.
Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el número de ataques por país y año desde el año 1900.

```python
paises = ['USA', 'AUSTRALIA', 'SOUTH AFRICA']
df_paises = df_attacks[df_attacks['country'].isin(paises)][df_attacks['year'] >= 1900]
df_paises.head(3)
```

```python
plt.figure(figsize=(10, 30))
fig4 = sns.countplot(data=df_paises, y = 'year', hue='country')
fig4.set(xlabel= 'Nº de ataques', ylabel= 'Años')
plt.show()
```

Sólo en los primeros años representados en la gráfica es Australia el país con más ataques con respecto a los otros dos. SIn embargo, a partir de 1957 la mayoría de ataques se dan en USA.

En el caso de África los registros son casi nulos en los primeros años, y se mantienen con pocas variaciones en los sucesivos.

### Ejercicio 5.

Realizad un boxplot donde se muestre los principales estadísticos del número de ataques por país.

```python
df_at_paises = df_attacks.groupby('country')['case_number'].count().reset_index(name='attacks')
df_at_paises
```

```python
plt.figure(figsize=(10,30))
fig5 = sns.boxplot(data=df_at_paises, x='attacks', y='country')
fig5.set_title('Nº de ataques por país')
plt.show();
```

Esta gráfica permite observar de manera muy evidente, como los tres países comentados en la representación anterior son los que registran más casos de ataques de tiburón.
En ella corroboramos que USA es el país donde más ataques se dan, seguido de Australia y Sudáfrica; el resto de países se mantienen en un rango de valores muy similar.
